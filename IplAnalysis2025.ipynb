```python
batsman_total=df.groupby('striker')['runs_of_bat'].sum().reset_index(name='Totalruns').sort_values('Totalruns',ascending=False)
```

```python
####Top10Batsman
top_batsman=batsman_total.reset_index().head(10)
top_batsman


```

```python
top_batsman['Rank']=top_batsman['Rank']= np.round(top_batsman['Totalruns'].rank(ascending=False),0)
top_batsman
```

```python
def random_color(no_of_colors):
    return['#'+''.join([np.random.choice(list('0123456789ABCDEF')) for _ in range(6)]) for _ in range(no_of_colors)]
random_color(6)
```

```python
plt.figure(figsize=(12,6))
colors = plt.cm.plasma(np.linspace(0, 1, len(top_batsman)))
sns.barplot(x='striker',y='Totalruns',data=top_batsman,palette=colors)
plt.title('IPL_2025_Top_10_Batsman')
plt.xlabel('Batsman Name')
plt.ylabel('Total runs')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python
df.head(1)
```

```python
##Best Performance Batsman in one inning
top_bats_score=df.groupby(['match_no','striker']).agg(Highest_score=('runs_of_bat','sum'),Balls_Fased=('runs_of_bat','count')).reset_index().sort_values('Highest_score',ascending=False).head(10)
```

```python
top_bats_score['Strike_Rate']=np.round((top_bats_score['Highest_score']/top_bats_score['Balls_Fased'])*100)
```

```python
top_bats_score
```

```python
####Top10 Batsman performance
top_strike_rate=top_bats_score.sort_values('Strike_Rate',ascending=False)
```

```python
top_strike_rate
```

```python



plt.figure(figsize=(12,6))
plt.bar(top_bats_score['striker'],top_bats_score['Highest_score'])
plt.title('IPL-2025 Highest Scorer Batsman')
plt.xlabel('Batsman Name')
plt.ylabel('Score')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12,6))
plt.bar(top_strike_rate['striker'],top_strike_rate['Strike_Rate'])
plt.title('IPL-2025 Highest Strikerate Batsman')
plt.xlabel('Batsman Name')
plt.ylabel('Strike rate')
plt.xticks(rotation=40)
plt.ylim(0, top_bats_score['Strike_Rate'].max() + 30) # Adds space at the top for labels
plt.tight_layout()
plt.show()
```

```python
####Top10 Bowlers
bowler_wicket=df[df['wicket_type']!='runout']
top_bowler=bowler_wicket.groupby(['bowler']).agg(Total_wicket=('wicket_type','count'),Run_giver=('runs_of_bat','sum')).reset_index().sort_values('Total_wicket',ascending=False).head(10)
```

```python
top_bowler
```

```python
plt.figure(figsize=(12,6))
plt.bar(top_bowler['bowler'],top_bowler['Total_wicket'])
plt.title('IPL-2025 Top Wicket Taker')
plt.xlabel('Bowler Name')
plt.ylabel('Total wicket')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()

```

```python
###Most wicket bowler in a inning
top_wicket=df.groupby(['match_no','bowler'])['wicket_type'].count().reset_index(name='Wicket').sort_values('Wicket',ascending=False)[:10]
top_wicket
```

```python
###Most Rungiver
bowler_run=df.groupby(['match_no','bowler'])['runs_of_bat'].sum().reset_index(name='Rungiven').sort_values('Rungiven',ascending=False)
bowler_run
```

```python
top_wicket_taker=pd.merge(top_wicket,bowler_run)
top_wicket_taker
```

```python
top_wicket_taker['performance']=top_wicket_taker['performance']=np.round(top_wicket_taker['Rungiven']/top_wicket_taker['Wicket'],2)
```

```python
top_wicket_rate=top_wicket_taker.sort_values('performance',ascending=True)
top_wicket_rate
```

```python
plt.figure(figsize=(12,6))
plt.bar(top_wicket_taker['bowler'],top_wicket_taker['Wicket'])
plt.title('IPL-2025 Highest Wicket Taker_One_Inning')
plt.xlabel('Bowler Name')
plt.ylabel('Wickets in a inning')
plt.xticks(rotation=40)
plt.ylim(0, 6)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(top_wicket_rate['bowler'],top_wicket_rate['performance'],marker='.',markersize=20)
plt.title('IPL-2025 Highest Wicket Taker Performance')
plt.xlabel('Bowler Name')
plt.ylabel('Runs per wicket in a inning')
plt.xticks(rotation=40)
plt.ylim(2,9)
plt.tight_layout()
plt.show()
```

```python
batsman_total=df.groupby('striker')['runs_of_bat'].sum().reset_index(name='Totalruns').sort_values('Totalruns',ascending=False)
```

```python
####Top10Batsman
top_batsman=batsman_total.reset_index()
top_batsman


```

```python
top_batsman['Rank']=top_batsman['Rank']= np.round(top_batsman['Totalruns'].rank(ascending=False),0)
top_batsman

```

```python
def random_color(no_of_colors):
    return['#'+''.join([np.random.choice(list('0123456789ABCDEF')) for _ in range(6)]) for _ in range(no_of_colors)]
random_color(6)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='striker',y='Totalruns',data=top_batsman,palette=random_color(len(top_batsman)))
plt.title('IPL2025 Top Batsman')
plt.xlabel('Batsman Name')
plt.ylabel('Total runs')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python
df.head(1)
```

```python
##Best Performance Batsman in one inning
top_bats_score=df.groupby(['match_no','striker'])['runs_of_bat'].sum().reset_index(name='Runs').sort_values('Runs',ascending=False)
top_bats_score
```

```python
balls_faced=df.groupby(['match_no','striker'])['runs_of_bat'].count().reset_index(name='Ball Faced')
balls_faced
```

```python
batsman_performance=pd.merge(top_bats_score,balls_faced)
batsman_performance
```

```python
batsman_performance['Strike rate']=batsman_performance['Strike rate']=np.round(batsman_performance['Runs']*100/batsman_performance['Ball Faced'],2)
```

```python
####Top10 Batsman performance
top_bats_performance=batsman_performance.head(10)
top_bats_performance
```

```python
plt.figure(figsize=(12,6))
plt.scatter(top_bats_performance['striker'],top_bats_performance['Strike rate'])
plt.title('IPL-2025 Highest Scorer Strikerate')
plt.xlabel('Batsman Name')
plt.ylabel('Strike rate')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python
top_strikerate=top_bats_performance.sort_values('Strike rate',ascending=False)
top_strikerate
```

```python
plt.figure(figsize=(12,6))
plt.scatter(top_strikerate['striker'],top_strikerate['Strike rate'])
plt.title('IPL2025 Top Strikerate Batsman')
plt.xlabel('Batsman Name')
plt.ylabel('Strike rate')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(top_bats_performance['striker'],top_bats_performance['Runs'],s=100,alpha=0.6)
plt.title('IPL-2025 Highest Run Scorer')
plt.xlabel('Batsman Name')
plt.ylabel('Runs')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python
####Top10 Bowlers
bowler_wicket=df[df['wicket_type']!='runout']
top_bowler=bowler_wicket.groupby(['bowler'])['wicket_type'].count().reset_index(name='Total wicket').sort_values('Total wicket',ascending=False).head(10)
top_bowler
```

```python
plt.figure(figsize=(12,6))
plt.plot(top_bowler['bowler'],top_bowler['Total wicket'],marker='*',markersize=20)
plt.title('IPL-2025 Top Wicket Taker')
plt.xlabel('Bowler Name')
plt.ylabel('Total wicket')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(12,6))
plt.plot(top_bowler['bowler'],top_bowler['Total wicket'],marker='*',markersize=20)
plt.title('IPL-2025 Top10 Bowler')
plt.xlabel('Bowler Name')
plt.ylabel('Wickets')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python
###Most wicket bowler in a inning
top_wicket=df.groupby(['match_no','bowler'])['wicket_type'].count().reset_index(name='Wicket').sort_values('Wicket',ascending=False)[:10]
top_wicket
```

```python
###Most Rungiver
bowler_run=df.groupby(['match_no','bowler'])['runs_of_bat'].sum().reset_index(name='Rungiven').sort_values('Rungiven',ascending=False)
bowler_run
```

```python
top_wicket_taker=pd.merge(top_wicket,bowler_run)
top_wicket_taker
```

```python
top_wicket_taker['performance']=top_wicket_taker['performance']=np.round(top_wicket_taker['Rungiven']/top_wicket_taker['Wicket'],2)
```

```python
top_wicket_taker=top_wicket_taker.sort_values('performance',ascending=False)
top_wicket_taker
```

```python
plt.figure(figsize=(12,6))
plt.plot(top_wicket_taker['bowler'],top_wicket_taker['Wicket'],marker='.',markersize=20)
plt.title('IPL-2025 Highest Wicket Taker')
plt.xlabel('Bowler Name')
plt.ylabel('Wickets in a inning')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(top_wicket_taker['bowler'],top_wicket_taker['performance'],marker='.',markersize=20)
plt.title('IPL-2025 Highest Wicket Taker Performance')
plt.xlabel('Bowler Name')
plt.ylabel('Runs per wicket in a inning')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```
